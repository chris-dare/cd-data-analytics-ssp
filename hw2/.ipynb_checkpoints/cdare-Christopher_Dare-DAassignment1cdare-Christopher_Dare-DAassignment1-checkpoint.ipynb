```python
def load_wind_gen():
#     loads and preprocesses intraday on-shore wind power generation measured every hour for one year
    date_parse_string = "%d/%m/%Y"
    df = pd.read_csv("data/WindGeneration.csv")
    df["Date"] = pd.to_datetime(
        df['Date'].astype(str),format=date_parse_string)
    return df
```

```python
# load wind generation data set
df = load_wind_gen()
df
```

```python
# visualize the time series
# tmp = df.groupby(pd.Grouper(key='Date'))
tmp = df
plt.figure(num=None, figsize=(12, 9), dpi=80, facecolor='w', edgecolor='k')
plt.xlabel("Time")
plt.ylabel("Wind Generation")
plt.title("Time Series plot of Wind Generation. Hourly data for jan 2014 - Jan 2015")
plt.plot(tmp["Date"], tmp["Wind Generation"])

# plt.legend()
```
